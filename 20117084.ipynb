# **QUESTION 1**

a.) Simulate 100 Poisson Random variables with mean = 14

b.) Take the average of these random variables

c.) Repeat step a.) and b.) 100000 times, everytime noting down the average.

d.) Draw a histogram of the 100000 averages that are stored with you.


```python
# importing important libraries
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

# n --> the number of times we need to repeat the steps
n = 100000 
mean = 14

# We append the averages of every 100 poisson random variables with mean = 14
# in a list called averages
PoissonAverages = []

for i in range(n):
  # We create a temporary average variable to store the current average
  tempAvrg = random.poisson(mean, size=100).mean()
  # then we append the temporary average to 'averages' list for plot
  PoissonAverages.append(tempAvrg)
```

```python
# Plotting the Histogram

plt.xlabel('Poisson Random Variable Averages', fontsize=16)
plt.ylabel('Average values', fontsize=16)
plt.title('Poisson Random Variables', fontsize=20)
plt.hist(PoissonAverages, bins=50, alpha=0.8)
plt.show()
```

# **QUESTION 2**

a.) Simulate 50 Normal Random variables with mean = 14, variance = 3

b.) Take the average of these random variables

c.) Repeat step a.) and b.) 100000 times, everytime noting down the average.

d.) Draw a histogram of the 100000 averages that are stored with you.

```python
# importing important libraries
import math
import numpy as np

# n --> the number of times we need to repeat the steps
n = 100000 
mean = 14
variance = 3
size1 = 50
stdDeviation = math.sqrt(variance)
NormVar = np.random.normal(mean, stdDeviation, size1)
NormVar.mean()

NormalAverages = []

for i in range(n):
  tempAvrg = np.random.normal(mean, stdDeviation, size1).mean()
  NormalAverages.append(tempAvrg)
```

```python
# Plotting the Histogram

plt.xlabel('Normal Random Variable Averages', fontsize = 16)
plt.ylabel('Average values', fontsize = 16)
plt.title('Normal Random Variable', fontsize = 20)
plt.hist(NormalAverages, bins = 50)
plt.show()
```

# **QUESTION 3**

a.) Simulate 50 random variables X which follow the following rule:

X = Y + Z

Y = Binomial random variable with n = 100, p = 0.8

Z = Uniform random variable with a = 40, b = 60

(The symbols have their standard meanings)

b.) Take the average of these random variables

c.) Repeat step a.) and b.) 100000 times, everytime noting down the average.

d.) Draw a histogram of the 100000 averages that are stored with you.

```python
n = 100
p = 0.8
size2 = 50
a = 40
b = 60
n = 100000

Y = np.random.binomial(n, p, size2)
Z = np.random.uniform(a, b, size2)
X = np.add(Y, Z)

Averages = []

for i in range(n):
  Y = np.random.binomial(n, p, size2)
  Z = np.random.uniform(a, b, size2)
  X = np.add(Y, Z)
  tempAvrg = X.mean()
  Averages.append(tempAvrg)
```

```python
plt.xlabel('Random Variable Averages', fontsize=16)
plt.ylabel('Average values', fontsize=16)
plt.title('Binomial and Uniform Random Variables', fontsize=20)
plt.hist(Averages, bins = 50)
plt.show()
```
