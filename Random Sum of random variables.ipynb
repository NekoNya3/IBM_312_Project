#To demonstrate the expected value and variance formula that we derived in class for a random variable which is given by a sum of "random number of random variables"

#N = Poisson random variable, hence E(N) = 50, Var(N) = 50

#P = 0, with 0.5 prob P = 1000, with 0.3 prob P = 5000, with 0.2 prob

#Thus, E(P) = 1300, Var(P) = (5000^2)0.2 + (1000^2)0.3 - 1300^2 = 3610000

#Thus, as per the derived expression, E(R) = 1300(50) = 65000 Var(R) = 50(3610000)+50((1300)(1300)) = 265000000

#We will now try to verify the results from simulation

```python
Revenue = []
for i in range(100000):
  N_cust = np.random.poisson(50)
  totalpurchase = 0
  for j in range(N_cust):
    rpurch = np.random.uniform(0,1)
    if rpurch<0.5:
      totalpurchase = totalpurchase + 0
    elif rpurch<0.8:
      totalpurchase = totalpurchase + 1000
    else:
      totalpurchase = totalpurchase + 5000
  Revenue.append(totalpurchase)  
```

```python
Revenue
```

```python
average = sum(Revenue)/len(Revenue)
average
```

```python
var = sum((x-average)**2 for x in Revenue) / len(Revenue)
var
```

```python
varest = var*(len(Revenue)/(len(Revenue)-1))
varest
```
